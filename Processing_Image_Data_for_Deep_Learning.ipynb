#### Resizing the image

```python
import PIL
from PIL import Image
```

```python
img = Image.open('/content/Random.jpg')
```

```python
resized_img = img.resize((200,200))
```

```python
resized_img.save('resized_img.jpg')
```

```python
img = mpimg.imread('/content/resized_img.jpg')
img_plot = plt.imshow(img)
plt.show()
```

#### Converting Rgb images to Grayscale image using OpenCV

```python
# importing OpenCV Library
import cv2
```

```python
img = cv2.imread('/content/Random.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(gray_img)
```

```python
gray_img.shape
```

```python
# cv2 imshow() will display the image, but this will be allowed in Google Colab.
# from google.colab.patches import cv2_imshow
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Displaying the Image
cv2_imshow(gray_img)
```

```python
# Perpose of this grayscale image
#1. To reduce the computation power
#2. To reduce the storage
#3. Avoid color image when color is not important for output like recognizing objects
#4. But in case we have to identify color of image that time we can't grayscale the image
```
